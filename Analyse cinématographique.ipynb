```python
from sqlalchemy import create_engine
import pandas as pd

# Connexion à la base
engine = create_engine("sqlite:///data/database.db")

# Requête SQL avec CTE (Common Table Expression)
query = """
WITH FilmPopularity AS (
  SELECT
    fc.category_id,
    f.film_id,
    f.title,
    c.name AS category_name,
    COUNT(i.inventory_id) AS inventory_count,
    ROW_NUMBER() OVER(
      PARTITION BY fc.category_id 
      ORDER BY COUNT(i.inventory_id) DESC
    ) AS popularity_rank
  FROM film f
  JOIN film_category fc ON f.film_id = fc.film_id
  JOIN category c ON fc.category_id = c.category_id
  JOIN inventory i ON f.film_id = i.film_id
  GROUP BY fc.category_id, f.film_id
)
SELECT
  film_id,
  title,
  category_name,
  inventory_count
FROM FilmPopularity
WHERE popularity_rank = 1;
"""


df_popular_films = pd.read_sql_query(query, con=engine)

# Affichage du résultat
df_popular_films

```

```python
from sqlalchemy import create_engine
import pandas as pd

# Connexion à la base SQLite
engine = create_engine("sqlite:///data/database.db")

# Ta requête complète avec sous-requête
query = """
SELECT
  fp.film_id,
  fp.title,
  fp.category_name,
  fp.inventory_count
FROM (
  SELECT
    fc.category_id,
    f.film_id,
    f.title,
    c.name AS category_name,
    COUNT(i.inventory_id) AS inventory_count,
    ROW_NUMBER() OVER(
      PARTITION BY fc.category_id 
      ORDER BY COUNT(i.inventory_id) DESC
    ) AS popularity_rank
  FROM film f
  JOIN film_category fc ON f.film_id = fc.film_id
  JOIN category c ON fc.category_id = c.category_id
  JOIN inventory i ON f.film_id = i.film_id
  GROUP BY fc.category_id, f.film_id
) AS fp
WHERE fp.popularity_rank = 1;
"""

# Lire le résultat dans un DataFrame
df_top_films = pd.read_sql_query(query, con=engine)

# Affichage
df_top_films
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(df_top_films['category_name'], df_top_films['inventory_count'])
plt.xlabel("Nombre d'exemplaires en stock")
plt.title("Film le plus populaire par catégorie")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Connexion à la base SQLite
engine = create_engine("sqlite:///data/database.db")

# Requête SQL corrigée (et multi-tables)
query = """
SELECT
  co.country,
  ci.city,
  COUNT(cu.customer_id) AS number_of_customers
FROM customer cu
JOIN address a ON cu.address_id = a.address_id
JOIN city ci ON a.city_id = ci.city_id
JOIN country co ON ci.country_id = co.country_id
GROUP BY co.country
ORDER BY number_of_customers DESC;
"""

# Exécution et affichage
df_customers_per_country = pd.read_sql_query(query, con=engine)
df_customers_per_country
```

```python
import matplotlib.pyplot as plt

# Agréger et trier les pays par nombre de clients (top 15)
df_country = df_customers_per_country.groupby("country")["number_of_customers"] \
                                     .sum() \
                                     .sort_values(ascending=False) \
                                     .head(15)

# Visualisation
plt.figure(figsize=(10, 6))
df_country.plot(kind='bar')
plt.title("Top 15 des pays avec le plus de clients")
plt.ylabel("Nombre de clients")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Connexion à la base SQLite
engine = create_engine("sqlite:///data/database.db")

# Requête SQL : nombre de locations par catégorie de film
query = """
SELECT
  cat.name AS category_name,
  COUNT(rent.rental_id) AS number_of_rentals
FROM rental rent
JOIN inventory inv ON rent.inventory_id = inv.inventory_id
JOIN film f ON inv.film_id = f.film_id
JOIN film_category fc ON f.film_id = fc.film_id
JOIN category cat ON fc.category_id = cat.category_id
GROUP BY cat.name
ORDER BY number_of_rentals DESC;
"""

# Exécution et récupération dans un DataFrame
df_rentals_by_category = pd.read_sql_query(query, con=engine)
df_rentals_by_category
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(df_rentals_by_category['category_name'], df_rentals_by_category['number_of_rentals'])
plt.xlabel("Nombre de locations")
plt.title("Nombre de locations par catégorie de film")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Connexion à la base SQLite
engine = create_engine("sqlite:///data/database.db")

# Requête SQL : dépense moyenne par client
query = """
SELECT
  AVG(total_spent) AS average_spent_per_client
FROM (
  SELECT
    SUM(payment.amount) AS total_spent
  FROM payment
  GROUP BY payment.customer_id
) AS customer_spending;
"""

# Exécution et affichage
df_avg_spent = pd.read_sql_query(query, con=engine)
df_avg_spent
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Connexion à la base SQLite
engine = create_engine("sqlite:///data/database.db")

# Requête SQL : clients classés par dépenses totales
query = """
SELECT
  cu.customer_id,
  cu.first_name || ' ' || cu.last_name AS customer_name,
  SUM(pa.amount) AS total_spent
FROM customer cu
JOIN payment pa ON cu.customer_id = pa.customer_id
GROUP BY cu.customer_id
ORDER BY total_spent DESC;
"""

# Exécution de la requête
df_top_spenders = pd.read_sql_query(query, con=engine)
df_top_spenders
```

```python
top_10 = df_top_spenders.head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(top_10['customer_name'], top_10['total_spent'])
plt.xlabel("Dépense totale (€)")
plt.title("Top 10 des clients les plus dépensiers")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Connexion à la base SQLite
engine = create_engine("sqlite:///data/database.db")

# Requête SQL : performance par magasin
query = """
SELECT
  s.store_id,
  COUNT(p.payment_id) AS number_of_transactions,
  ROUND(SUM(p.amount), 2) AS total_revenue,
  ROUND(AVG(p.amount), 2) AS average_revenue_per_rental
FROM store s
JOIN staff sta ON s.store_id = sta.store_id
JOIN payment p ON sta.staff_id = p.staff_id
GROUP BY s.store_id
ORDER BY total_revenue DESC;
"""

# Exécution
df_store_revenue = pd.read_sql_query(query, con=engine)
df_store_revenue
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df_store_revenue['store_id'].astype(str), df_store_revenue['total_revenue'])
plt.title("Revenu total par magasin")
plt.xlabel("ID du magasin")
plt.ylabel("Revenu total (€)")
plt.tight_layout()
plt.show()
```
