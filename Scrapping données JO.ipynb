❓**Question 5 :** A l'aide de boucles et de listes, récupérez la [table des villes hôtes](https://fr.wikipedia.org/wiki/Jeux_olympiques_d%27%C3%A9t%C3%A9#Villes_h%C3%B4tes).

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1. Récupération de la page
url = "https://fr.wikipedia.org/wiki/Jeux_olympiques_d%27%C3%A9t%C3%A9#Villes_h%C3%B4tes"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 2. Trouver tous les tableaux
tables = soup.find_all("table", class_="wikitable")

# 3. Identifier la table des éditions (souvent la première "wikitable center")
# On va prendre celle avec beaucoup de lignes et les bonnes colonnes
for table in tables:
    headers = table.find_all('th')
    if headers and "Olympiade" in headers[0].get_text():
        host_table = table
        break

# 4. Récupérer les lignes
rows = host_table.find_all("tr")

# 5. Initialiser une liste pour stocker les données
data = []

# 6. Boucle sur les lignes (en ignorant l'entête multiple)
for row in rows[2:]:
    cols = row.find_all(["td", "th"])
    cols_text = [col.get_text(strip=True) for col in cols]
    data.append(cols_text)

# 7. Définir les noms de colonnes (à adapter selon les colonnes visibles dans le tableau)
colonnes = ["Olympiade", "Année", "Hôte", "Dates", "Nations", "Athlètes Total", "Hommes", "Femmes", "Sports", "Épreuves"]

# 8. Créer le DataFrame
df = pd.DataFrame(data, columns=colonnes[:len(data[0])])  # Ajuste si certaines colonnes sont absentes

# 9. Afficher les 5 premières lignes
print(df.head())
```

❓**Question 6 :** Vérifiez que chaque colonne est bien formatée (c'est-à-dire qu'il n'y a pas de contenu manquante ou anormal comme des liens ou des balises HTML, des parenthèses avec des numéros, etc...). Vérifiez également que chaque colonne a le bon type :

- Olympiade : String
- Année : Integer
- Pays hôte : String (à convertir en liste)
- Date : String
- Nations : Integer
- Athlètes (toutes sous-colonnes) : Integer
- Sports : Integer
- Épreuves : Integer

Nettoyez la donnée scrappée.

```python
import pandas as pd
import numpy as np
import re
```

```python
def nettoyer_texte(cell):
    if pd.isna(cell):
        return cell
    cell = str(cell)
    cell = re.sub(r'\[\d+\]', '', cell)           # Supprime [1], [2], ...
    cell = re.sub(r'\([^)]*\)', '', cell)         # Supprime (1), (texte), ...
    cell = re.sub(r'<.*?>', '', cell)             # Supprime balises HTML
    return cell.strip()

# Appliquer le nettoyage sur tout le DataFrame
df_clean = df.applymap(nettoyer_texte)
# Colonnes à convertir en int
colonnes_int = ["Année", "Nations", "Athlètes Total", "Hommes", "Femmes", "Sports", "Épreuves"]

for col in colonnes_int:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
print(df_clean.dtypes)
print(df_clean.head())
```
